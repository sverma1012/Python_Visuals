### Project: The Legalisation of Same-Sex Marriage

#### <b> Description </b>
Creating a time slider visual using Python to show when a country the legal status of same-sex marriages.

#### <b> Part Two </b>
Combining the external data that contains the legalisation status with the csv file that contains geospatial information.

##### by Sneha Verma

-------------------

```python
# Import packages.

import pandas as pd
import numpy as np
```

```python
# Load external data.

legalisation_status = pd.read_csv("marriage-same-sex-partners-equaldex.csv")
print(legalisation_status.head())
```

```python
# Load geospatial data.

geospatial_data = pd.read_csv("countries_geospatial.csv")
print(geospatial_data.head())
```

-------------

Join both datasets.

```python
# Join the geospatial data on the legalisation to get the lat and long for each country.

country_legalisation = legalisation_status.merge(
    geospatial_data, left_on= "Entity", right_on="Country", how="left"
)

print(country_legalisation.head())
```

---------------

```python
# Export dataframe as csv.

country_legalisation.to_csv("country_legalisation_full.csv", index=False)
```
